```python
import numpy as np
```

```python
seq1 = pd.read_csv('./trajectory/up1.csv')
seq2 = pd.read_csv('./trajectory/up2.csv')
```

```python
seq1 = np.array(seq1)
seq2 = np.array(seq2)
print(seq1.shape)
print(seq2.shape)
```

```python
seq1 = np.append(seq1, seq2, axis=0)
np.set_printoptions(suppress=True)
seq1
```

```python
x_max = max(seq1[:,0])
x_min = min(seq1[:,0])
y_max = max(seq1[:,1])
y_min = min(seq1[:,1])
z_max = max(seq1[:,2])
z_min = min(seq1[:,2])
print(x_min, x_max, y_min, y_max, z_min, z_max)
```

```python
seq1.shape
seq1 = seq1.reshape(2,35,3)
seq1
```

```python
seq1.shape[0]
```

```python
tmp = np.zeros((2,35,3))
```

```python

for i in range(seq1.shape[0]):
    for j in range(seq1.shape[1]):
        tmp[i,j,0] = (seq1[i,j,0]-x_min)/(x_max-x_min)
        tmp[i,j,1] = (seq1[i,j,1]-y_min)/(y_max-y_min)
        tmp[i,j,2] = (seq1[i,j,2]-z_min)/(z_max-z_min)
        
tmp
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, RepeatVector, TimeDistributed
```

```python
hidden_layer = 256
n_step = 35
features = 3
```

```python
model = Sequential()
model.add(LSTM(hidden_layer, activation='relu', input_shape=(n_step, features)))
model.add(RepeatVector(n_step))
model.add(LSTM(hidden_layer, activation='relu', return_sequences=True))
model.add(TimeDistributed(Dense(3)))
```

```python
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(tmp, tmp, batch_size=2, epochs=300)
```

```python
yhat = model.predict(tmp)
yhat
```

```python
pred = yhat
for i in range(yhat.shape[0]):
    for j in range(yhat.shape[1]):
        pred[i,j,0] = pred[i,j,0]*(x_max-x_min)+x_min
        pred[i,j,1] = pred[i,j,1]*(y_max-y_min)+y_min
        pred[i,j,2] = pred[i,j,2]*(z_max-z_min)+z_min
pred
```

```python
error = np.zeros((seq1.shape[0], seq1.shape[1], seq1.shape[2]))
for i in range(error.shape[0]):
    for j in range(error.shape[1]):
        for k in range(error.shape[2]):
            error[i,j,k] = seq1[i,j,k]-pred[i,j,k]
error
```
