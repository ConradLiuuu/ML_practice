```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense, RepeatVector, TimeDistributed
import pandas as pd
```

```python
traj1 = pd.read_csv('traj11.csv')
print(traj1)
print(traj1.shape)
```

```python
traj2 = pd.read_csv('traj22.csv')
print(traj2)
print(traj2.shape)
```

```python
n_step = len(traj1)
print(n_step)
```

```python
samples = 1 # batch_size
features = 3 # n_input
hidden_layer = 256
epochs = 500
```

```python
seq1 = np.array(traj1)
seq2 = np.array(traj2)
seq1
```

```python
seq1 = seq1.reshape(samples, n_step, features)
max_x = np.max(seq1[:,:,0])
min_x = np.min(seq1[:,:,0])
max_y = np.max(seq1[:,:,1])
min_y = np.min(seq1[:,:,1])
max_z = np.max(seq1[:,:,2])
min_z = np.min(seq1[:,:,2])
print(max_x, min_x, max_y, min_y, max_z, min_z)
```

```python
tmp_x = []
tmp_y = []
tmp_z = []
for i in seq1[:,:,0]:
    tmp_x.append((i-min_x)/(max_x-min_x))
tmp_x = np.array(tmp_x)
#print(tmp_x)

for i in seq1[:,:,1]:
    tmp_y.append((i-min_y)/(max_y-min_y))
tmp_y = np.array(tmp_y)

for i in seq1[:,:,2]:
    tmp_z.append((i-min_z)/(max_z-min_z))
tmp_z = np.array(tmp_z)

for i in range(35):
    seq1[:,i,0] = tmp_x[0,i]
    seq1[:,i,1] = tmp_y[0,i]
    seq1[:,i,2] = tmp_z[0,i]
#for i in range(len(tmp_x)):
    #seq1[:,i,0] = tmp_x[i,0]
print(seq1)
#print(seq1[:,2,0])
```

```python
model = Sequential()
model.add(LSTM(hidden_layer, activation='relu', input_shape=(n_step, features)))
model.add(RepeatVector(n_step))
model.add(LSTM(hidden_layer, activation='relu', return_sequences=True))
model.add(TimeDistributed(Dense(3)))
```

```python
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
model.fit(seq1, seq1, epochs=epochs, verbose=2)
```

```python
seq2 = seq2.reshape(samples, n_step, features)
yhat = model.predict(seq1, verbose=0)
print(yhat)
print(seq1)
```

```python
for i in range(35):
    yhat[:,i,0] = yhat[:,i,0]*(max_x-min_x)+min_x
    yhat[:,i,1] = yhat[:,i,1]*(max_y-min_y)+min_y
    yhat[:,i,2] = yhat[:,i,2]*(max_z-min_z)+min_z
print(yhat)
```

```python
error = np.zeros((1,35,3))
#print(error)
for i in range(35):
    error[:,i,0] = seq_act[:,i,0]-yhat[:,i,0]
    error[:,i,1] = seq_act[:,i,1]-yhat[:,i,1]
    error[:,i,2] = seq_act[:,i,2]-yhat[:,i,2]
np.set_printoptions(suppress=True)
print(error)
```
